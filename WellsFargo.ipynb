```python
# creating an empty list to store the links

links = []



# creating loop to extract links of all job openings

for num in range(1,168):
        
    url = f'https://www.wellsfargojobs.com/en/jobs/?page={num}#results'
    driver.get(url)
#     driver.maximize_window()
    sleep(8)

    for element in driver.find_elements_by_xpath('//a[@class="stretched-link"]'):
        links.append(element.get_attribute("href"))
        print(element.get_attribute("href"))
        
        
#     Pagination code not in use for this case                
#     next_button = driver.find_element_by_xpath('//a[@aria-label="Go to next page of results"]')
#     next_button.click()
#     print(f'Page {num} is scraped')
#     sleep(3)



    
        
```

```python
# creating an empty dictionary to store the data

info = []
counter = 0


# extracting data from all the individual links

for link in links:
    
    driver.get(link)
    sleep(5)
    
     
    for title in driver.find_elements_by_xpath('//h1'):
        jobtitle = title.text
        print(jobtitle)
        
      
    for detail in driver.find_elements_by_xpath('//div[@class="text"]/ul[@class="list-inline job-meta"]'):
        details = detail.text
        
        
    for desc in driver.find_elements_by_xpath('//article'):
        description = desc.text
        
    
    info.append({'Job_Title':jobtitle, 'Details':details, 'Description':description, 'Links':link})
    
    counter = counter+1
    print(counter)
       
```

```python
data = pd.DataFrame.from_dict(info)
```
