```python
df_san
```

```python
df_san.to_excel('Sankey_PDL.xlsx')
```

```python
unique_source_target = list(pd.unique(df_san[['source','target']].values.ravel('K')))
```

```python
unique_source_target
```

```python
mapping_dict = {k: v for v, k in enumerate(unique_source_target)}
```

```python
mapping_dict
```

```python
df_san['source'] = df_san['source'].map(mapping_dict)
df_san['target'] = df_san['target'].map(mapping_dict)
```

```python
df_san.head(10)
```

```python
san_dict = df_san.to_dict(orient='list')
san_dict
```

```python
unique_source_target
```

```python
fig = go.Figure(data=[go.Sankey(
      node=dict(
      pad = 15,
      thickness = 20,
      line = dict(color = 'black', width = 0.5),
      label = unique_source_target,
      color = 'blue'
      ),
    
      link = dict(
      source = san_dict['source'],
      target = san_dict['target'],
      value = san_dict['value']
        
      
      )
)])
```
