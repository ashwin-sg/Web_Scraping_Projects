```python
# provide chromedriver path

PATH = "C:\Webdrivers\Chromedriver 69\chromedriver.exe"


# Open Chrome and Access Linkedin login site

driver = webdriver.Chrome(PATH)
sleep(2)

print('Web driver is launched successfully')
```

#### Extracting a data

```python
# creating empty lists to store the data

job_title = []
reference = []
location = []


# creating pagination

for num in range(0,49,6):
    
    url = f'https://manpowergroup.avature.net/talentsolutionsrpo/SearchJobs/?folderRecordsPerPage=6&folderOffset={num}'
    driver.get(url)
    sleep(15)
    
    for title in driver.find_elements_by_xpath('//h3[@class="article__header__text__title article__header__text__title--6"]/a'):
        job_title.append(title.text)
        
#     for ref in driver.find_elements_by_xpath('')
#         reference.append(ref.text)
      
    for loc in driver.find_elements_by_xpath('//span[@class="subtitle__item"]/span'):
        location.append(loc.text)
 
    
    
```
