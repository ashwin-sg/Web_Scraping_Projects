#### Extracting a data

## EMEA

```python
# creating empty list to store the links


links = []


# creating pagination


for num in range(0,7,6):
    
    url = f'https://workmyway.avature.net/ManpowerGroupTalentSolutions/SearchJobs/?jobRecordsPerPage=6&jobOffset={num}'
    driver.get(url)
    sleep(15)
    
    for title in driver.find_elements_by_xpath('//h3/a[@class="link"]'):
        links.append(title.get_attribute("href")) 
        print(title.get_attribute("href"))
```

```python
links
```

```python
# creating an empty list to store the data

info = []


# extracting data from all the individual links

for link in links:
    
    driver.get(link)
    sleep(10) 
       
    
    for title in driver.find_elements_by_xpath('//h2[@class="banner__text__title banner__text__title__title--1"]'):
        jobtitle = title.text
        print(jobtitle)
        
    for detail in driver.find_elements_by_xpath('//article[@class="article article--details regular-fields--cols-2Z regular-fields-label--inline js_collapsible"]'):
        details = detail.text
        print(detail.text)
   
    for detail in driver.find_elements_by_xpath('//article[@class="article article--details  js_collapsible"]'):
        description = detail.text

    info.append({
        
        'Job_Title': jobtitle,
        'Details'  : details,
        'Description' : description,
        'Links' : link
        
    })
    
    sleep(5)
```

```python
data_emea = pd.DataFrame.from_dict(info)
```
