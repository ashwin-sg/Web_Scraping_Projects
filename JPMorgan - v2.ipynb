```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(PATH, options=options)
```

#### Extracting data

```python
# loading url

url = 'https://jpmc.fa.oraclecloud.com/hcmUI/CandidateExperience/en/sites/CX_1001/requisitions?keyword='
driver.get(url)
driver.maximize_window()
sleep(10)
```

```python
# creating pagination

for num in range(1,14):

        element = driver.find_element_by_tag_name('body')
        element.send_keys(Keys.PAGE_DOWN)
            
        print(f'Page {num} is scraped')
        sleep(5) 
```

```python
for num in range(14,575):
      
        next_button = driver.find_element_by_xpath('//button[@class="search-results-load-more-btn"]')
        next_button.click()
        sleep(9)
            
        print(f'Page {num} is scraped')
     
```

```python
# creating empty list to store the links


links = []

    
for title in driver.find_elements_by_xpath('//a[@class="joblist-tile"]'):
    links.append(title.get_attribute("href"))
    print(title.get_attribute("href"))
    
    
    

     
    
```

```python
links_part2 = links[2375:4674]
```

```python
# creating an empty list to store the data

info = []
counter = 2374


# extracting data from all the individual links

for link in links_part2:
    
    driver.get(link)
    sleep(5) 
       
    
    for title in driver.find_elements_by_xpath('//h2[@class="job-title"]/span[@data-bind="text: job().title"]'):
        jobtitle = title.text
        print(title.text)

    for loc in driver.find_elements_by_xpath('//span[@class="primary-location"]'):
        location = loc.text

    for detail in driver.find_elements_by_xpath('//div[@class="job-details-list"]'):
        details = detail.text

    for des in driver.find_elements_by_xpath('//div[@class="details"]'):
        description = des.text


    info.append({

        'Job_Title': jobtitle,
        'Location' : location,
        'Details'  : details,
        'Description' : description,
        'Links' : link

    })
    
    counter = counter+1
    print(counter)
```

```python
data = pd.DataFrame.from_dict(info)
```
