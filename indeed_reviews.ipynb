```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import csv 
from lxml import etree
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options


print("Finish importing packages")
```

```python
# provide chromedriver path

PATH = "C:\Webdrivers\Chromedriver 69\chromedriver.exe"


# Open Chrome and Access Linkedin login site

driver = webdriver.Chrome(PATH)
sleep(3)

print('Web driver is launched successfully')
```

```python
df = pd.read_excel('links.xlsx')
df.head()
```

```python
links = []

links = df['url']
```

```python
links
```

```python
# collecting data for ROS the links
    
driver.get('https://www.indeed.com/cmp/Randstad-Sourceright')
driver.maximize_window()
sleep(10)
    
for element in driver.find_elements_by_xpath('//div[@class="css-1maxdqz eu4oa1w0"]'):
    name = element.text
    print(name)
```

```python
# create a list to store information

info = []

# applying loop to extract information

for link in links:
    
    driver.get(link)
    driver.maximize_window()
    sleep(10)
    
    
    for element in driver.find_elements_by_xpath('//div[@class="css-19rjr9w e1wnkr790"]'):
        name = element.get_attribute("innerHTML")
        print(name)
    
    for element in driver.find_elements_by_xpath('//span[@class="css-s5cg6m e1wnkr790"]'):
        score = element.get_attribute("innerHTML")
        print(score)
      
    for element in driver.find_elements_by_xpath('//span[@class="css-htn3vt e1wnkr790"]'):
        total = element.get_attribute("innerHTML")
    
    for element in driver.find_elements_by_xpath('//div[@class="css-1maxdqz eu4oa1w0"]'):
        other = element.text
        
    
    info.append({
        
        'Name' : name,
        'Total Score': score,
        'Total Rating' : total,
        'Other' : other,
        'Links' : link
        
    })
    
    sleep(5)
```

```python
data = pd.DataFrame.from_dict(info)
```
