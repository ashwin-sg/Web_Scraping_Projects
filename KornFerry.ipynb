#### Extracting a data

```python
# creating empty list to store the links


links = []


# creating pagination

for num in range(0,301,100):
    
    url = f'https://kornferry.tal.net/vx/lang-en-GB/mobile-0/appcentre-1/brand-4/xf-a9ce85135610/candidate/jobboard/vacancy/3/adv/?start={num}'
    driver.get(url)
    sleep(15)
    
    for title in driver.find_elements_by_xpath('//td/a[@class="subject"]'):
        links.append(title.get_attribute("href"))   
```

```python
len(links)
```

```python
links
```

```python
links_2 = links[155:]
```

```python
# creating empty lists to store the data

info = []

# extracting data from all the individual links

for link in links_2:
    
    driver.get(link)
    
    sleep(10)
        
   
    
    for title in driver.find_elements_by_xpath('//h1[@class="section"]'):
        jobtitle = title.text
       
    for req in driver.find_elements_by_xpath('//div[@id="form_label_93306_1"]'):
        requisition = req.text
     
    for cn in driver.find_elements_by_xpath('//div[@id="form_field_140006_1"]'):
        country = cn.text
    
    for ct in driver.find_elements_by_xpath('//div[@id="form_field_151365_1"]'):
        city = ct.text

    for desc in driver.find_elements_by_xpath('//div[@id="form_field_value_81251_1"]'):
        description = desc.text
        
        
    info.append({
        
        'Job_Title': jobtitle,
        'Requisition'  : requisition,
        'City' : city,
        'Country' : country,
        'Description' : description,
        'Links' : link
        
    })
    
    sleep(5)
```

```python
data_2 = pd.DataFrame.from_dict(info)
```
